Comparation for base and increvental with graphs

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
datadir='data/'
picdir='pictures/'
dataset='D2'
```

```python
increm_df=pd.read_csv(datadir+dataset+'increm.csv')
base_df=pd.read_csv(datadir+dataset+'base.csv')
if dataset=='D1':
    d=pd.date_range(start='2017-04-08',end='2017-11-01',  freq='2w').strftime('%Y-%m-%d')
else:
    d=pd.date_range(start='2019-10-15',end='2020-09-30',  freq='2w').strftime('%Y-%m-%d')
```

```python
l=np.arange(len(base_df))
plt.figure(figsize=(12,9))
plt.plot(increm_df.aucTot, c='b', ls='--', marker='o',lw=3, markersize=12, label='Incremental')

plt.plot(base_df.aucTot, c='r', ls='--', marker='o',lw=3, markersize=12,label='Base')

plt.xlabel('Last day for step', size=16)
plt.ylabel('AUC-ROC', size=16)
plt.yticks(size=16)
plt.xticks(l,d[5:], rotation=60, size=16)
plt.grid()
plt.legend(fontsize=16)
plt.tight_layout()
plt.savefig(picdir+dataset+'auc_compare.png', dpi=300, format='png' )
plt.show()
```

```python
l=np.arange(len(base_df))
plt.figure(figsize=(12,9))
plt.plot(increm_df.aucA, c='g', ls='--', marker='o', lw=3, markersize=12,label='Class A' )
plt.plot(increm_df.aucB, c='r', ls='--', marker='o',lw=3, markersize=12,label='Class B')
plt.plot(increm_df.aucTot, c='blue', ls='--', marker='o',lw=3, markersize=12,label='Both classes')
plt.xlabel('Last day for step', size=16)
plt.ylabel('AUC-ROC', size=16)
plt.yticks(size=16)
plt.xticks(l,d[5:], rotation=60, size=16)
plt.grid()
plt.legend(fontsize=16)
plt.tight_layout()
plt.savefig(picdir+dataset+'auc2clincr.png', dpi=300, format='png' )
plt.show()
```

```python
l=np.arange(len(base_df))
plt.figure(figsize=(12,9))

plt.plot(increm_df.CE, c='blue', ls='--', marker='o',lw=3, markersize=12, label='Incremental')
plt.plot(base_df.CE, c='red', ls='--', marker='o',lw=3, markersize=12, label='Base')

plt.xlabel('Last day for step', size=16)
plt.ylabel('Median CE', size=16)
plt.yticks(size=16)
plt.xticks(l,d[5:], rotation=60, size=16)
plt.grid()
plt.legend(fontsize=14)
plt.tight_layout()
plt.savefig('D1CE_compare_base.png', dpi=300, format='png'  )
plt.show()
```
